```python
import pandas as pd
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
```

```python
customers = pd.read_csv("/kaggle/input/datasets/sahil8989/customer-features/customer_features.csv")
```

```python
rfm = customers[["recency", "frequency", "monetary"]]
```

```python
scaler = StandardScaler()

rfm_scaled = scaler.fit_transform(rfm)
```

```python
kmeans = KMeans(

    n_clusters=4,
    random_state=42

)

customers["cluster"] = kmeans.fit_predict(rfm_scaled)
```

```python
cluster_map = {

    0: "High-value loyal customers",
    1: "At-risk customers",
    2: "New buyers",
    3: "Price-sensitive users"

}

customers["segment"] = customers["cluster"].map(cluster_map)
```

```python
customers[["customer_id", "segment"]].to_csv(

    "customer_segments.csv",
    index=False

)

print("Customer segmentation completed")
```
